```python
import pandas as pd
import numpy as np
import ast
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_df = merged_data[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']].copy()
```

```python
print(merged_data.shape)
print(merged_data.head())
```

```python
movies_df = merged_data[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']].copy()
```

```python
print(movies_df.isnull().sum())
movies_df.dropna(inplace=True)
```

```python
movies_df = movies_df.dropna(subset=['overview'])
```

```python
print(movies_df.isnull().sum())
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

movies_df['genres'] = movies_df['genres'].apply(convert)
movies_df['keywords'] = movies_df['keywords'].apply(convert)
```

```python
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

movies_df['cast'] = movies_df['cast'].apply(convert_cast)
```

```python
def fetch_director(obj):
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            return [i['name']]
    return []

movies_df['crew'] = movies_df['crew'].apply(fetch_director)
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x: x.split())
```

```python
# Combine all features into a single string per row
movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']

# Convert the list to a single string (with space-separated words)
movies_df['tags'] = movies_df['tags'].apply(lambda x: " ".join(x))

# Convert to lowercase
movies_df['tags'] = movies_df['tags'].apply(lambda x: x.lower())

# Check the result
print(movies_df[['title', 'tags']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# TF-IDF vectorization on 'tags' column
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies_df['tags'])

# Calculate cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
similarity_df = pd.DataFrame(cosine_sim, index=movies_df['title'], columns=movies_df['title'])

# Show top 5x5 portion
similarity_df.iloc[:5, :5]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(similarity_df.iloc[:5, :5], cmap='coolwarm')  # Top 15 movies
plt.title("Cosine Similarity")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
